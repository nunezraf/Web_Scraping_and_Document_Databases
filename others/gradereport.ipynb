```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import pandas as pd 
```

```python
# Read HTML from file
filepath = os.path.join("report6411.html")
with open(filepath) as file:
    html = file.read()
```

```python
# Create a Beautiful Soup object
soup = bs(html, 'lxml')
```

```python
# Extract title text
title = soup.title.text
print(title)
```

```python
print(html)
```

```python
students_name = soup.find_all('b')
for name in students_name:
    print(name.text)
```

```python
# getting data from students 
students = soup.find_all('p', attrs={'class': 'bigFont'})
# results_data= results.text.strip() 
students 
```

```python
#getting data from classes 
classes = soup.find_all('td', attrs={'class': 'bigFont'})
# results_data= results.text.strip() 
classes
```

```python
#getting the grades 
grades = soup.find_all('td', attrs={'a': 'a'})
# grades_df= pd.DataFrame(grades)
for grade in grades:
    print (grades[10],)
```
